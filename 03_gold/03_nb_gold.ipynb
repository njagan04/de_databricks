```python
from pyspark.sql.functions import avg, count, hour, col


df = spark.table("silver.transformed_data")
```

Aggregation by pickup ZIP

```python

df_zip_metrics = (
    df.groupBy("pickup_zip")
      .agg(
          avg("fare_amount").alias("avg_fare"),
          avg("trip_distance").alias("avg_trip_distance"),
          avg("trip_duration_minutes").alias("avg_trip_duration"),
          count("*").alias("total_trips")
      )
)

display(df_zip_metrics)
```

```python
spark.sql("CREATE SCHEMA IF NOT EXISTS gold")
df_zip_metrics.write.mode("overwrite").saveAsTable("gold.nyctaxi_zip_metrics")
```

Aggregation by pickup HOUR 

```python
df_hourly = (
    df
    .withColumn("pickup_hour", hour(col("pickup_datetime")))
    .groupBy("pickup_zip")
    .agg(
        avg("fare_amount").alias("avg_fare"),
        avg("trip_distance").alias("avg_trip_distance"),
        avg("trip_duration_minutes").alias("avg_trip_duration"),
        count("*").alias("total_trips")
    )
)

display(df_hourly)

df_hourly.write.mode("overwrite").saveAsTable("gold.nyctaxi_hourly_metrics")
```

Overall trip stats

```python
df_kpi = (
    df.agg(
        avg("fare_amount").alias("overall_avg_fare"),
        avg("trip_distance").alias("overall_avg_distance"),
        avg("trip_duration_minutes").alias("overall_avg_duration"),
        count("*").alias("total_trips")
    )
)

display(df_kpi)

df_kpi.write.mode("overwrite").saveAsTable("gold.nyctaxi_kpi")
```
