```python
from pyspark.sql.functions import (
    col, year, month, dayofmonth,
    avg, sum, count, abs, max, min,
    current_date, trunc, add_months, when,
    to_date, last_day, rank, row_number
)
from pyspark.sql.window import Window
```

```python
order_silver = spark.table("02_silver_catalog.silver.order")
invoice_silver = spark.table("02_silver_catalog.silver.invoice")
estimate_silver = spark.table("02_silver_catalog.silver.estimate")
customer_survey_silver = spark.table("02_silver_catalog.silver.customer_survey")
store_silver = spark.table("02_silver_catalog.silver.store")
ns_budget_silver = spark.table("02_silver_catalog.silver.ns_budget")

```

### Dimensional Tables

```python
dim_store = store_silver.select(
    "store_id",
    "store_name",
    "city",
    "state",
    "manager_id",
    "manager_name",
    "store_type",
    "opened_year"
).dropDuplicates()

dim_technician = order_silver.select(
    "technician_id",
    "technician_name",
    "store_id"
).dropDuplicates()

dim_estimator = estimate_silver.select(
    "estimator_id",
    "estimator_name"
).dropDuplicates()

dim_date = order_silver.select(
    col("vehicle_in_date").alias("date")
).union(
    order_silver.select(col("delivery_date").alias("date"))
).union(
    invoice_silver.select(col("invoice_date_only").alias("date"))
).dropDuplicates()

dim_date = dim_date.withColumns({
    "year": year(col("date")),
    "month": month(col("date")),
    "day": dayofmonth(col("date"))
})
```
